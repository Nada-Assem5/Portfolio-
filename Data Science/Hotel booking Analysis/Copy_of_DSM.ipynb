# **cleaning**

```python
data.drop(columns='company',inplace=True)
# removes unnecessary data
```

```python
data.dropna(subset=['children','country'],inplace=True)
#removes missing values (Nulls).
```

```python
data['agent'].fillna(555,inplace=True)
# handles missing values.
```

```python
data = data [data['adults' ]!=0]
#remove wrong values
```

```python
data.drop_duplicates(inplace=True)
#remove duplicates
```

```python
data = data[((data['market_segment']=='Direct') | (data['market_segment']=='Groups')) & (data['distribution_channel']=='Direct') |
      ((data['market_segment']=='Corporate') & (data['distribution_channel']=='Corporate')) |
      ((data['market_segment']=='Online TA') | (data['market_segment']=='Offline TA/TO')) & (data['distribution_channel']=='TA/TO')]
# Remove the rows with inconvenient data
```

```python
data =data[((data['is_canceled'] == 0) & (data['reservation_status'] == 'Check-Out')) | ((data['is_canceled'] == 1) & (data['reservation_status'] == 'Canceled'))]
# Remove the rows with inconvenie data
```

```python
data = data[~((data['adr']==0) & (data['market_segment']!='Complementary'))]
# Remove the rows with inconvenient data
```

```python
Q1=data['lead_time'].quantile(0.25)
Q3=data['lead_time'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
data=data[(data['lead_time']>=lower_bound) & (data['lead_time']<=upper_bound)]
# remove the outliers from Lead_Time
```

```python
Q_1=data['adr'].quantile(0.25)
Q_3=data['adr'].quantile(0.75)
IQR=Q_3-Q_1
lower_bound=Q_1-1.5*IQR
upper_bound=Q_3+1.5*IQR
data=data[(data['adr']>=lower_bound) & (data['adr']<=upper_bound)]
# remove the outliers from ADR
```

```python
sns.boxplot(data['lead_time'])
plt.title('Lead Time outliers check')
plt.show()
# check deleting the outliers
```

```python
sns.boxplot(data['adr'])
plt.title('ADR outliers check')
plt.show()
# check deleting the outliers
```

# **Analysis & Visualization**

```python
plt.hist(data['hotel'],bins = 3 , edgecolor ='Black' , color='green' , alpha=0.5)
plt.title('Hotel Distribution')
plt.xlabel('Hotel')
plt.ylabel('Frequency')
plt.show()
```

The graph shows the **distribution of hotel types** in the dataset, divided into two main categories:  
1. Resort Hotel
2. City Hotel  

**Observations and Interpretation:**
1. **City Hotel** has a significantly higher frequency compared to **Resort Hotel**.  
   - The City Hotel accounts for approximately **62%** of the data.  
   - The Resort Hotel accounts for approximately **38%** of the data.  

**Percentage Calculation:  **
  - City Hotel  **45,000**  
  - Resort Hotel  **28,000**  
- Total = 45,000 + 28,000 = 73,000  

- **City Hotel**: \( \frac{45,000}{73,000} \times 100 \approx 61.6\% \)  **62%**  
- **Resort Hotel**: \( \frac{28,000}{73,000} \times 100 \approx 38.4\% \)  **38%**  




City Hotels dominate the dataset, representing around **62%** of the total observations, while Resort Hotels represent around **38%**. This suggests that City Hotels are more frequent, possibly due to their popularity for business trips, urban tourism, or general travel purposes.

```python
# Plot cancellation rate
sns.countplot(x='is_canceled', data=data['is_canceled'].to_frame())
plt.title('Cancellation Rate')
plt.xlabel('Is Canceled (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python
# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create a count plot for reservation status by hotel type
plt.figure(figsize=(12, 6))
data1 = data[['hotel', 'reservation_status']]
sns.countplot(data=data1, x='hotel', hue='reservation_status', palette='Set2')

# Add titles and labels
plt.title('Reservation Status by Hotel Type', fontsize=16)
plt.xlabel('Hotel Type', fontsize=14)
plt.ylabel('Number of Reservations', fontsize=14)
plt.legend(title='Reservation Status', fontsize=12)
plt.xticks(rotation=0)  # Rotate x labels if needed
plt.show()
```

```python
# Calculate cancellation rates by hotel type
cancellation_rate = data.groupby('hotel')['is_canceled'].mean() * 100  # Convert to percentage
cancellation_rate = cancellation_rate.reset_index()
cancellation_rate.columns = ['Hotel Type', 'Cancellation Rate (%)']
```

```python
# Plot cancellation rates by hotel type
plt.figure(figsize=(8, 5))
sns.barplot(x='Hotel Type', y='Cancellation Rate (%)', data=cancellation_rate, palette='YlGnBu')
plt.title('Cancellation Rate by Hotel Type', fontsize=16)
plt.xlabel('Hotel Type', fontsize=14)
plt.ylabel('Cancellation Rate (%)', fontsize=14)
plt.xticks(rotation=0)
plt.show()
```

The graph illustrates the reservation status for two hotel types: Resort Hotel and City Hotel.

The reservation statuses are divided into two categories:

* Check-Out (successful stays)
* Canceled (canceled reservations)
                        _______________________________________________
* City Hotels experience more reservations overall, but they also have a higher
cancellation rate (29%) compared to Resort Hotels (21%).
* This could suggest that City Hotels are more likely to attract short-term or business travelers, who may be more prone to cancel their plans.
* Resort Hotels, on the other hand, might attract vacationers with more solid travel plans, leading to fewer cancellations.

```python
# Count reservation status per month
status_per_month = data.groupby(['arrival_date_month', 'reservation_status']).size().unstack(fill_value=0)

# Plot reservation status per month
plt.figure(figsize=(12,6))
status_per_month.plot(kind='bar', stacked=True , color=['#66785F','#91AC8F','#4B5945'])
plt.title('Reservation Status Per Month', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Reservations', fontsize=14)
plt.legend(title='Reservation Status', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

**Manage High-Season Demand (July & August):**

Observation: The months of August and July have the highest number of reservations, particularly check-outs.
Action:
Increase staff availability during these months to manage the surge in guests.
Offer promotions or early booking discounts to spread demand across other months.
Ensure room maintenance and inventory (like amenities) are well-prepared in advance.
**Reduce Cancellations in Off-Season (January & February):**

Observation: The months of January and February show fewer total reservations with a relatively higher proportion of cancellations.
Action:
Introduce flexible cancellation policies or discounts for early bookings to encourage customers.
Launch targeted marketing campaigns, such as winter vacation packages or family deals.
Partner with travel agencies to promote these quieter months.
**Focus on Marketing for Mid-Tier Months (April, June, September):**

Observation: April, June, and September show moderate reservation numbers.
Action:
Promote mid-tier months with seasonal packages, like spring or early autumn deals.
Create special offers for long stays or group bookings during these months.
*Monitor Cancellation Trends Year-Round:**

Observation: Cancellations occur every month, with some months (like May and August) having notable levels.
Action:
Introduce non-refundable rates at a discount for customers confident in their plans.
Offer incentives for customers to reschedule instead of canceling, such as vouchers or discounts for future stays.
**Improve Guest Retention During December:**

Observation: December has lower check-outs but moderate cancellations.
Action:
Create festive packages for holidays, targeting families and leisure travelers.
Collaborate with local attractions to make December stays more appealing.
